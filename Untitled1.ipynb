## Data preprocessing

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the provided dataset
file_path = r"C:\Users\parul\Downloads\Book1.xlsx"
data = pd.read_excel(file_path)
```

```python
# Initial data exploration
data.shape
print(data.head())  # Display the first few rows of the dataset
print(data.info())  # Get information about the dataset, including missing values
```

```python
# Handle missing data (if any)
data.dropna(inplace=True)  # Remove rows with missing values
```

```python
# Handle outliers (if any)
# We can use statistical methods or visualization to identify and remove outliers
# For example, using the IQR method:

# Handle outliers using the IQR method
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Align the data before performing comparisons
data = data[~((data < (Q1 - 1.5 * IQR)).any(axis=1) | (data > (Q3 + 1.5 * IQR)).any(axis=1))]
```

```python
# Encoding categorical variables
# For categorical variables, we can use label encoding or one-hot encoding
label_encoder = LabelEncoder()
from sklearn.preprocessing import LabelEncoder

# Encode the 'Gender' column
data['COUNTRY'] = label_encoder.fit_transform(data['COUNTRY'])

# Encode the 'Location' column
data['COMPANYNAME'] = label_encoder.fit_transform(data['COMPANYNAME'])

# Encode the 'Churn' column as it's categorical
data['ProductCategoryText.SHORT_DESCR'] = label_encoder.fit_transform(data['ProductCategoryText.SHORT_DESCR'])
```
